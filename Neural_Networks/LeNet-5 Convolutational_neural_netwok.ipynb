### Train a Convolutional Neural Network

#### Convolutional Neural network Architecture

```python
class conv_nn(torch.nn.Module):
    
    def __init__(self):
        
        super(conv_nn, self).__init__()
        
        # convolutation layer
        # input_chanels = 1
        # output: 6
        # trainable weights: Weights + bias
        # 5 (filter size) x 5 (filter size) x 1 (chanels) x 6 (num_filters) + 6 (num_filters)
        # trainable weights = 156 
        self.conv1 = torch.nn.Conv2d(1, 6, kernel_size = 5, stride = 1, padding = 2, bias = True)
        
        # average pooling
        self.avgpooling = torch.nn.AvgPool2d(kernel_size = 2, stride = 2)

        # activation function
        self.act = torch.nn.Tanh()

        # 2 convoluional layer
        # Connect only 10 out of 16 feaures maps
        self.conv2 = torch.nn.Conv2d(6, 16, kernel_size = 5, stride = 1, padding = 0, bias = True)
        
        # fully connected convolutional layer
        # input: 16 x 4 x 4
        # output: 120
        self.fcc = torch.nn.Linear(16*5*5, 120, bias = True)
        
        # fully connect layer
        self.fc = torch.nn.Linear(120, 84, bias = True)
        
        # Fully connected softmax output layer
        self.soft_fc = torch.nn.Sequential(
        
            torch.nn.Linear(84, 10, bias = True),
            torch.nn.Softmax()
            
        )
        
        # Dropout layer
        self.dropout = torch.nn.Dropout2d(p = 0.5)
        
        
    # forward method
    def forward(self, x):
        
        # size of minibatch
        input_size = x.size(0)
        
        # 1 convolution
        x = self.act(self.conv1(x))
        
#         print('out 1 conv: ', x.shape)
        
        # 1 avg pooling
        x = self.avgpooling(x)
        
#         print('out 1 pooling', x.shape)
        
        # 2 convolution
        x = self.act(self.conv2(x))
        
#         print('out 2 conv: ', x.shape)
        
        # 2 pooling
        # Reduce dimension
        x = self.avgpooling(x)
        
#         print('out 2 pooling: ', x.shape)
        
        # flatten
        x = x.view(input_size, -1)
        
#         print('flatten out: ', x.shape)
        
        # fully connected convolutional layer
        x = self.act(self.fcc(x))
        
#         print('out fcc: ', x.shape)
        
        # Fully connect layer
        x = self.act(self.fc(x))
        
#         print('out fc: ', x.shape)
        
        x = self.dropout(x)
        
        x = self.soft_fc(x)
        
#         print('out softmax final layer: ', x.shape)
          
        return x
    
    # Method for apply only first layers to input
    # layer is the layer to apply (if it's 1, so only apply 1 layer (conv + pool), 2 apply 2 conv with their 2 pool)
    def first_filters(self, x, layer = 1):

        x = self.act(self.conv1(x))
        x = self.avgpooling(x)

        if layer == 2:
            
            # 2 convolution
            x = self.act(self.conv2(x))

            # 2 pooling
            # Reduce dimension
            x = self.avgpooling(x)
        
        return x
```

#### Set configurations for NN

```python
# define net
net = conv_nn()
```

### GPU Implementation

```python
# GPU implementation
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")

# flag to use GPU
use_gpu = True if torch.cuda.is_available() else False

# Allocate net in GPU
net.to(device)

print(device)
```

### Define optimizer

```python
# Define optimizer
optimizer = torch.optim.Adam(net.parameters(), lr=1e-4)
criterion = torch.nn.NLLLoss()
```

### Train NN

#### Se realiza training considerando 100 epochs. En cada epoch se obtiene el error sobre el conjunto de training y sobre el conjunto de validation, graficandose utilizando la libreria hiddenlayer.

Se destaca: 

1) Training se ejecuta en GPU.

2) Se implementa early stopping considerando que si error de validacion es mayor que error de training se comienza a tener sobreajuste, sin embargo en todo momento se tiene esta situación, en donde el error en trainig siempre es menor. Frente a esto, se plantea como posible implementación, analizar si un ambos errores se diferencian en cierto porcentaje (con el objetivo de definir que si ambos se alejan demasiado, entonces probablmente se tenga un sobreajuste).

```python
# Hiddenlayer objects to track metrics
import hiddenlayer as hl

history1 = hl.History()
history2 = hl.History()

canvas1 = hl.Canvas()


# number of epochs
epochs = 100

for epoch in range(epochs):
    
    epoch_loss_train, epoch_acc_train = 0.0, 0.0
    
    # Training
    for data, label in train_loader:
        
        # If use GPU, so allocate data in device (CUDA)
        if use_gpu:
            
            data, label = data.cuda(), label.cuda()
        
        # make predictions
        prediction = net(data)
        
        # zero gradient = not acumulating value
        optimizer.zero_grad()  
        
        # get loss value
        loss = criterion(prediction, label)  
        
        # Update error
        epoch_loss_train += loss.item()
        epoch_acc_train += (prediction.argmax(dim=1) == label).sum().item()
               
        # backprop
        loss.backward()
        
        # Update parameters
        optimizer.step()
    
    
#         print(epoch_acc)

    # Validation
    
    # loss
    epoch_loss_valid, epoch_acc_valid = 0.0, 0.0
    
    for data, label in valid_loader:
        
        if use_gpu:
            
            data, label = data.cuda(), label.cuda()
            
        prediction = net(data)
        
        loss = criterion(prediction, label)  
        
        epoch_loss_valid += loss.item()
        
        epoch_acc_valid += (prediction.argmax(dim=1) == label).sum().item()  
        
#         print(epoch_acc)
    
    history1.log(epoch, loss_valid=epoch_loss_valid/len(valid_idx), accuracy_valid=epoch_acc_valid/len(valid_idx))
    history2.log(epoch, loss_train=epoch_loss_train/len(train_idx), accuracy_train=epoch_acc_train/len(train_idx))
    
    with canvas1: # So that they render together
        canvas1.draw_plot([history1["loss_valid"], history2["loss_train"]])
        canvas1.draw_plot([history1["accuracy_valid"], history2["accuracy_train"]])

    # Early stopping
    # However, It has all time, valid loss is greater than train, so this implementation does not work in this case
    if epoch_loss_train < epoch_loss_valid and epoch > 5:
        
        print('Early stopping in epoch: ', epoch)
        
#         break
        
```

#### Analisis

Se observa que error de training es menor para todo epoch comparado con validation. Esto genera que el modelo se esté ajustando en mayor magnitud a los datos de trainig. Con respecto al erro, se tiene tambien que éste disminuye en función del epoch para el trainig, lo que afirma que en cada epoch, el modelo se asjuta mas al conjunto de training. Con respecto al error de validación, se tiene este disminuye pero no con la misma tasa que trainig.

Al observar accuracy de ambos conjuntos de datos, se tiene que para ambos conjuntos, el accuracy aumenta. Sin embargo el accuracy es mayor para trainig, lo que confirma que el modelo se sobreajusta a los datos de trainig.

Desde lo anterior, se tiene sobreajuste sobre el dato de training, por lo que quizás una posible solución sea agregar dropout a otras capas (actualmente se tiene una capa con dropout).

### Evaluate obtained model with test data

```python
# Without gradient
with torch.no_grad():
    
    
    loss_test = 0.0
    predictions = np.empty(len(mnist_test))
    labels = np.empty(len(mnist_test))
    
    # For compute accuracy
    total = 0
    
    # minibatches for validation dataset
    for data, label in test_loader:
        
        # get prediction ussing trained model
        prediction = net(data)
       
        loss = criterion(prediction, label) 
        
        loss_test += loss
        
        # get prediction (max of prediction vector)
        _, predicted = torch.max(prediction.data, 1)
        
        predictions[total: total + 256] = predicted.numpy()
        
        labels[total: total + 256] = label
        
        # Get total of instances
        total += label.size(0)
```

### Get metrics of model

```python
from sklearn.metrics import confusion_matrix
from sklearn.metrics import f1_score
from sklearn.metrics import accuracy_score

print('confusion matrix: \n', confusion_matrix(labels, predictions))
print('f1 score: ', f1_score(labels, predictions, average='macro'))
print('accuracy score: ', accuracy_score(labels, predictions))
print('Test loss: ', loss_test.item())
```

#### Analysis

Al observas las métricas obtenidas sobre el conjunto de prueba, se tiene que accuracy de 98, 12 [%]. También se tiene f1 score de 98,10 [%], valor similar a accuracy ya que se debe estar analizando datos con clases balanceadas.
